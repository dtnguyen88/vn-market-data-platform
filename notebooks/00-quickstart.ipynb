# vnmarket SDK Quickstart

A minimal walkthrough of the `vnmarket` Python SDK. Run cell-by-cell after `gcloud auth application-default login`.

**Prerequisites:** Phase 01 + Phase 02 deployed in staging; ADC configured.

```python
import vnmarket as vm

client = vm.Client(env="staging")
client.project, client.bucket
```

## Daily OHLCV

```python
df = client.daily(["VNM", "VIC"], start="2024-01-02", end="2024-01-31")
df.head()
```

## Intraday ticks

```python
ticks = client.ticks("VNM", "2024-01-15").collect()
ticks.head()
```

## L2 snapshot

```python
book = client.l2_at("VNM", "2024-01-15T10:30:00+07:00")
{k: v[:5] for k, v in book.items()}
```

## Indices

```python
vnindex = client.index("VNINDEX", start="2024-01-02", end="2024-01-31", daily=True)
vnindex.head()
```

## Factors (BQ-backed view)

```python
factors = client.factors(start="2024-01-02", end="2024-01-31")
factors.head()
```

## Gap awareness

```python
client.gaps(stream="ticks")
```

## Custom SQL

```python
client.sql("SELECT 1 AS x")
```
